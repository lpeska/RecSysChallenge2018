```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("analyseDistribution2.csv", sep=",", header=0, index_col=0)
df = df.loc[df.LENGTH > 1]
df.head()
```

```python
df.describe()
```

```python
df.acousticness = df.acousticness * 1 / np.log(df.LENGTH)
df.danceability = df.danceability * 1 / np.log(df.LENGTH)
df.speechiness = df.speechiness * 1 / np.log(df.LENGTH)
df.energy = df.energy * 1 / np.log(df.LENGTH)
df.liveness = df.liveness * 1 / np.log(df.LENGTH)
df.instrumentalness = df.instrumentalness * 1 / np.log(df.LENGTH)
df["mode"] = df["mode"] * 1 / np.log(df.LENGTH)
df.loudness = df.loudness * 1 / np.log(df.LENGTH)
df.tempo = df.tempo * 1 / np.log(df.LENGTH)
df.duration_ms = df.duration_ms * 1 / np.log(df.LENGTH)
df.ALL = df.ALL * 1 / np.log(df.LENGTH)
df.REFINED = df.REFINED * 1 / np.log(df.LENGTH)
df.AUTHOR = df.AUTHOR * 1 / np.log(df.LENGTH)
```
