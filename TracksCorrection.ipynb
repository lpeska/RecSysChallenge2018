```python
for i in range(df.shape[0]):
    for j in range(1,df.shape[1]+1):
        existing = []
        if len(df[j].iloc[i]) == 37:
            existing.append(df[j].iloc[i])
        else:
            sl = df[j].iloc[i][0:37]
            if sl in existing:
                df[j].iloc[i] = " spotify:track:06rSBLNJ1H79orHkkQKr0P"
            else:
                df[j].iloc[i] = sl

df.iloc[1030:1045]
```
