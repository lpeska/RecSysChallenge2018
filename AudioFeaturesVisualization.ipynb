```python
import pandas as pd
df = pd.read_csv("trackAcoustics.csv", sep=",", header=None)
df.rename(index=str, columns={0: "id", 
                              1: "acousticness", 
                              2: "danceability", 
                              3: "loudness", 
                              4: "valence", 
                              5: "tempo", 
                              6: "time_signature", 
                              7: "speechiness", 
                              8: "energy", 
                              9: "liveness", 
                              10: "mode", 
                              11: "instrumentalness", 
                              12: "duration_ms"}, inplace=True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df[["acousticness", "danceability", "speechiness", "energy", "liveness", "instrumentalness", "mode", "loudness", "tempo", "duration_ms"]])
df[["acousticness", "danceability", "speechiness", "energy", "liveness", "instrumentalness", "mode", "loudness", "tempo", "duration_ms"]] = scaler.transform(df[["acousticness", "danceability", "speechiness", "energy", "liveness", "instrumentalness", "mode", "loudness", "tempo", "duration_ms"]])
df.head()
```

```python
df[["id","acousticness", "danceability", "speechiness", "energy", "liveness", "instrumentalness", "mode", "loudness", "tempo", "duration_ms"]].to_csv("trackAudioStd.csv", sep=",", float_format="%.6f")
```
